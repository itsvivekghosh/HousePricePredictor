```python
import sweetviz
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train=pd.read_csv("House Datasets/train.csv")
test = pd.read_csv("House Datasets/test.csv")
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)
train.head().T
```

```python
my_report = sweetviz.analyze([train, "Train"],target_feat='SalePrice')
```

```python
my_report.show_html('Report.html')
```

### Comparing two dataframes (e.g. Test vs Training sets)

```python
my_report1 = sweetviz.compare([train, "Train"], [test, "Test"], "SalePrice")
```
