```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", None)
```

```python
df=pd.read_csv("House Datasets/test.csv")
df.head()
```

```python
plt.figure(figsize=(20, 8))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
for feature in df.columns:
    if(np.round(df[feature].isnull().mean()*100, 2) > 0.0):
        print("{} has {}% of null values i.e {} out of {}".format(
            feature,
            np.round(df[feature].isnull().mean()*100, 2),
            (df[feature].isnull().sum()),
            len(df)
        ))
```

```python
df["MSZoning"]=df["MSZoning"].fillna(df["MSZoning"].mode()[0])
```

```python
df['LotFrontage']=df['LotFrontage'].fillna(df['LotFrontage'].mean())
df.drop(["Alley", "Id"], axis=1, inplace=True)
```

```python
df["Utilities"]=df["Utilities"].fillna(df.Utilities.mode()[0])
```

```python
df["MasVnrArea"]=df.MasVnrArea.fillna(df.MasVnrArea.mean())
df["MasVnrType"]=df.MasVnrType.fillna(df["MasVnrType"].mode()[0])
```

```python
bsmt_features=[x for x in df.columns if('Bsmt' in x)]
bsmt_features
```

```python
for feature in bsmt_features:
    
    if(df[feature].dtypes == 'object'):
        
        df[feature]=df[feature].fillna(df[feature].mode()[0])
    else:
        
        df[feature]=df[feature].fillna(df[feature].mean())
```

```python
df.Exterior1st, df.Exterior2nd = df.Exterior1st.fillna(df.Exterior1st.mode()[0]), df.Exterior2nd.fillna(df.Exterior2nd.mode()[0])
```

```python
for feature in df.columns:
    
    if(np.round(df[feature].isnull().mean()*100, 2) > 0.0):
        print("{} has {}% of null values i.e {} out of {}".format(
            feature,
            np.round(df[feature].isnull().mean()*100, 2),
            (df[feature].isnull().sum()),
            len(df)
        ))
```

```python
for feature in df.columns:
    
    if(np.round(df[feature].isnull().mean()*100, 2) > 0.0):
        print(feature, "==>", df[feature].dtypes)
```

```python
year_feature = [x for x in df.columns if('Yr' in x) or ('Year' in x)]
year_feature
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    df[feature]=df["YrSold"]-df[feature]
    
df[year_feature].head()
```

```python
df["Functional"]=df.Functional.fillna(df.Functional.mode()[0])
df["KitchenQual"]=df.KitchenQual.fillna(df.KitchenQual.mode()[0])
```

```python
df["FireplaceQu"]=df.FireplaceQu.fillna(df.FireplaceQu.mode()[0])
```

```python
garage_features = [x for x in df.columns if('Garage' in x)]
garage_features
```

```python
for feature in garage_features:
    
    if(df[feature].dtypes == 'object'):
        
        df[feature]=df[feature].fillna(df[feature].mode()[0])
    else:
        
        df[feature]=df[feature].fillna(df[feature].mean())
```

```python
df.SaleType=df.SaleType.fillna(df.SaleType.mode()[0])
```

```python
for feature in df.columns:
    
    if(np.round(df[feature].isnull().mean()*100, 2) > 0.0):
        print("{} has {}% of null values i.e {} out of {}".format(
            feature,
            np.round(df[feature].isnull().mean()*100, 2),
            (df[feature].isnull().sum()),
            len(df)
        ))
```

```python
for feature in df.columns:
    
    if(np.round(df[feature].isnull().mean()*100, 2) > 0.0):
        print(feature, "==>", df[feature].dtypes)
```

```python
df.drop(['PoolQC', 'Fence', 'MiscFeature'], axis=1, inplace=True)
```

```python
plt.figure(figsize=(20, 8))
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
le=LabelEncoder()
sc=StandardScaler()
```

```python
num_features=[x for x in df.columns if df[x].dtypes != 'O']
cat_features=[x for x in df.columns if df[x].dtypes == 'O']
```

```python
for feature in cat_features:
    
    df[feature]=le.fit_transform(df[feature].astype('str'))
```
